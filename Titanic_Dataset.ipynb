##Selección & Filtrado


```python
# Selección de columnas
df['age']                 # Serie
df[['pclass', 'fare', 'survived']]      # Sub-DataFrame

# Filtrado condicional
df[df['age'] > 25]
df.query('age > 25 and fare >= 50')
df[df['sex'].isin(['male', 'female'])]

# Selección por posición / etiqueta
df.loc[0:5, 'survived':'fare']   # por label
df.iloc[0:5, [1, 3, 4]]              # por índice

# Selección con patrón (columnas que empiezan con 'p')
df.filter(regex='^p')
```

##Limpieza de Datos

```python
# Manejo de nulos
df.dropna(subset=['age'])
df['age'] = df['age'].fillna(df['age'].median())
df.fillna(df.ffill())
df.infer_objects(copy=False)

# Duplicados
df.drop_duplicates(subset=['passengerId'], keep='last')

# Tipos de datos
#df['fecha'] = pd.to_datetime(df['fecha']) Este dataset no tiene ninguna fecha
df['who'] = df['who'].astype('category')

# Renombrar y reemplazar
df.rename(columns={'pclass': 'passengerClass'}, inplace=True)
df['survived'] = df['survived'].replace({0: 'No', 1: 'Yes'})

df.head() # Para verque los cambios de haya reflejado
```

##Agrupación & Agregación

```python
# Group By básico
df.groupby('sex')['fare'].mean() # Tafica promedio agrupada por sexo
```

```python
# Múltiples agregaciones
df.groupby('passengerClass').agg(
    edad_promedio = ('age', 'mean'),
    tarifa_max    = ('fare', 'max'),
    total_pasajeros = ('passengerId', 'count')
)
```

```python
# Pivot Table
pd.pivot_table(df,  # Suma (sum) todas las tarifas de todos los pasajeros que tengan la misma classe (class)
    index='embark_town',    # y que embarcaron en la misma ciudad
    values='fare',
    columns='class',
    aggfunc='sum'
)
```

```python
# Crosstab
pd.crosstab(df['embark_town'], df['who'])         # Es un contador de los valores en comun de filas y columnas
```

##Joins & Merge

```python
# Dataframe creado para poder hacer la operacion al no exitir dos tablas
class_desc = pd.DataFrame({
    'passengerClass': [1, 2, 3],
    'class_name': ['Primera', 'Segunda', 'Tercera']
})

# Merge (SQL-style)
pd.merge(df, class_desc,
    left_on='passengerClass',
    right_on='passengerClass',
    how='left'          # left | right | outer | cross
)
```

```python
# Concatenar DataFrames
pd.concat([df, class_desc], axis=0)  # vertical
pd.concat([df, class_desc], axis=1)  # horizontal

# Join por índice
df.join(class_desc.set_index('passengerClass'), on='passengerClass', how='inner')
```

##Tranformaciones

```python
# Apply: funciones custom
df['who_upper'] = df['who'].apply(lambda x: str(x).upper())
df['rango_tarifa'] = df['fare'].apply(
    lambda x: 'alto' if x > 50 else 'bajo'
)

# --- Window Functions ---
df['rolling_fare'] = df['fare'].rolling(7).mean() # Media móvil de tarifa
df['fare_cum_sum'] = df['fare'].cumsum()          # Suma acumulada de ingresos
df['fare_pct_change'] = df['fare'].pct_change()   # Cambio porcentual entre registros
df['rank_fare'] = df['fare'].rank(ascending=False)# Ranking de quién pagó más

df
```

```python
def limpiar_datos(df):
    return df.dropna(subset=['age'])

def agregar_features(df):
    df['FamilySize'] = df['sibsp'] + df['parch'] + 1
    return df

def filtrar_outliers(df):
    return df[df['fare'] > 300]

# Pipe: encadenar transformaciones
result = (df
    .pipe(limpiar_datos)
    .pipe(agregar_features)
    .pipe(filtrar_outliers)
)

result
```

##Exportar & Optimizar

```python
# Exportar
df.to_csv('Titanic_procesado.csv', index=False)
df.to_parquet('Titanic_procesado.parquet')

# Optimización de memoria
df['passengerClass'] = df['passengerClass'].astype('int8')
df['embarked'] = df['embarked'].astype('category')
df['sex'] = df['sex'].astype('category')

# Procesamiento por chunks (archivos grandes)
for chunk in pd.read_csv('Titanic.csv', chunksize=100):
    # Edad promedio de cada 100 datos
    print(chunk['age'].mean())

# Parquet con particiones
df.to_parquet('output_Titanic/', partition_cols=['alive'])

df_alive = pd.read_parquet('output_Titanic/alive=yes/')
print(df_alive)
```
